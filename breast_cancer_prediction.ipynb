```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("data.csv")
```

```python
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.dropna(axis=1, inplace=True)
```

```python
X = df.drop(["diagnosis","id"], axis=1)
```

```python
X.info()
```

```python
y = df["diagnosis"]
```

```python
y
```

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.model_selection import cross_val_score
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled
```

```python
le = LabelEncoder()
y_encoded = le.fit_transform(y)
y_encoded
```

```python
X_train,X_test,y_train,y_test = train_test_split(X_scaled,y_encoded, test_size = 0.2, random_state=42)
```

```python
k_values = range(1,10)
```

```python
accuracy_results = []
```

```python
for k in k_values:
    neigh = KNeighborsClassifier(n_neighbors= k)
    scores = cross_val_score(neigh, X_train, y_train, cv=5)
    accuracy_results.append(np.mean(scores))
```

```python
accuracy_results
```

```python
np.argmax(accuracy_results)
```

```python
best_k = k_values[np.argmax(accuracy_results)]
best_k
```

```python
best_accuracy = max(accuracy_results)
best_accuracy
```

```python
best_neigh = KNeighborsClassifier(n_neighbors=best_k)
```

```python
best_neigh.fit(X_train,y_train)
```

```python
y_pred = best_neigh.predict(X_test)
```

```python
y_pred
```

```python
acc = accuracy_score(y_test, y_pred)
```

```python
acc
```

```python
cf = confusion_matrix(y_test, y_pred)
cf
```

```python
tn, fp, fn, tp = cf.ravel()
```

```python
precision = tp / (tp + fp)
```

```python
precision
```

```python
recall = tp / (tp + fn)
```

```python
recall
```
